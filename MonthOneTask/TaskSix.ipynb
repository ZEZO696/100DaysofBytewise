```python
from sklearn.datasets import load_iris
import pandas as pd

iris = load_iris()
data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

data.head(2)
```

```python
def dataset_info(data):
    n_samples, n_features = data.shape
    return n_samples, n_features

n_samples, n_features = dataset_info(data)
print(f"Number of samples: {n_samples}")
print(f"Number of features: {n_features}")
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, iris.target, test_size=0.2, random_state=42)

X_train.shape, X_test.shape
```

```python
import matplotlib.pyplot as plt

plt.hist(data.iloc[:, 0], bins=20, edgecolor='k')
plt.title(f"Distribution of {data.columns[0]}")
plt.xlabel(data.columns[0])
plt.ylabel("Frequency")
plt.show()
```

```python
numbers = data[iris.feature_names[0]].head(10).tolist()

mean_value = sum(numbers) / len(numbers)
print(f"List of 10 numbers from the Iris dataset: {numbers}")
print(f"Mean of the list: {mean_value}")
```

```python
import numpy as np

def compute_statistics(numbers):
    count = len(numbers)
    mean = np.mean(numbers)
    median = np.median(numbers)
    std_dev = np.std(numbers)

    return {
        "count": count,
        "mean": mean,
        "median": median,
        "standard_deviation": std_dev
    }

stats = compute_statistics(numbers)
print(stats)
```
